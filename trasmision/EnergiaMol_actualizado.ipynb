```python
import time
import pandas as pd
from elasticsearch import Elasticsearch
from datetime import datetime

# Cargar CSV de datos reales
csv_path = "/Users/guillermofernandez/Desktop/descargas_celec/combinados_por_anio/Produccion_CSR_2024.csv"
df = pd.read_csv(csv_path)

# ✅ Corregimos formato de fecha: día/mes/año
df["Fecha"] = pd.to_datetime(df["Fecha"], dayfirst=True)
df = df[["Fecha", "EnergiaMol"]].dropna().sort_values("Fecha").reset_index(drop=True)

# Configuración de Elasticsearch
es = Elasticsearch(hosts=["http://localhost:9200"])

# Verificar conexión
if es.ping():
    print("✅ Conexión exitosa a Elasticsearch")
else:
    print("❌ Error al conectar con Elasticsearch")
    exit()

# Nombre del índice
indice = "energiamol-2024"

# Crear índice si no existe
if not es.indices.exists(index=indice):
    es.indices.create(index=indice)
    print(f"📁 Índice '{indice}' creado.")

# Función para crear documento desde el DataFrame
def crear_documento(row):
    return {
        "ubicacion": "CELEC - Planta Molino",
        "energia_mol": round(row["EnergiaMol"], 2),
        "timestamp": row["Fecha"].isoformat()
    }

# Bucle de envío
try:
    for i, row in df.iterrows():
        doc = crear_documento(row)
        res = es.index(index=indice, document=doc)
        print(f"[{i+1}/{len(df)}] Documento insertado → ID: {res['_id']}")
        print(doc)
        time.sleep(1)  # ⏱️ Espera 1 segundo por cada hora simulada
except KeyboardInterrupt:
    print("\n🛑 Transmisión detenida manualmente.")
```
