```python
import time
import pandas as pd
from elasticsearch import Elasticsearch
from datetime import datetime

# Cargar CSV de datos reales
csv_path = "/Users/guillermofernandez/Desktop/descargas_celec/combinados_por_anio/Produccion_CSR_2024.csv"
df = pd.read_csv(csv_path)

# ‚úÖ Corregimos formato de fecha: d√≠a/mes/a√±o
df["Fecha"] = pd.to_datetime(df["Fecha"], dayfirst=True)
df = df[["Fecha", "EnergiaSop"]].dropna().sort_values("Fecha").reset_index(drop=True)

# Configuraci√≥n de Elasticsearch
es = Elasticsearch(hosts=["http://localhost:9200"])

# Verificar conexi√≥n
if es.ping():
    print("‚úÖ Conexi√≥n exitosa a Elasticsearch")
else:
    print("‚ùå Error al conectar con Elasticsearch")
    exit()

# Nombre del √≠ndice
indice = "energiasop-2024"

# Crear √≠ndice si no existe
if not es.indices.exists(index=indice):
    es.indices.create(index=indice)
    print(f"üìÅ √çndice '{indice}' creado.")

# Funci√≥n para crear documento desde el DataFrame
def crear_documento(row):
    return {
        "ubicacion": "CELEC - Planta Sopladora",
        "energia_mol": round(row["EnergiaSop"], 2),
        "timestamp": row["Fecha"].isoformat()
    }

# Bucle de env√≠o
try:
    for i, row in df.iterrows():
        doc = crear_documento(row)
        res = es.index(index=indice, document=doc)
        print(f"[{i+1}/{len(df)}] Documento insertado ‚Üí ID: {res['_id']}")
        print(doc)
        time.sleep(1)  # ‚è±Ô∏è Espera 1 segundo por cada hora simulada
except KeyboardInterrupt:
    print("\nüõë Transmisi√≥n detenida manualmente.")
```
